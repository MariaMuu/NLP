```python
# Importing important libraries

import requests
from bs4 import BeautifulSoup
```

```python
# store url
url = 'https://www.gutenberg.org/cache/epub/39263/pg39263-images.html'
```

```python
# make the request
response = requests.get(url)
```

```python
# extract html from response
html = response.text
print(html)
```

```python
# get a more readabe text
soup = BeautifulSoup(html, 'html5lib')
```

```python
# get the book's title but in a good written way

soup.title.text.strip()
```

```python
# get the text out of the soup

text = soup.get_text()
print(text)
```

```python
paragraphs = soup.find_all('p')
paragraphs
```

```python
# get the text out of the soup

# initialize an empty list to store the text from each paragraph
only_paragraph_text = []

# iterate through the ResultSet of paragraphs
for paragraph in paragraphs:
    # Get the text from each individual paragraph and append it to the list
    only_paragraph_text.append(paragraph.get_text())

# Join all the paragraph texts together with a newline character in between
text = "\n".join(only_paragraph_text)

print(text)
```

Extract words from the text with NLP

```python
# import regex library

import re
```

```python
# find all words in the book and print the first 50.

words = re.findall('\w+', text)
print(words[:50])
```

```python
# Initialize new list
lower_words = []

# Loop through list tokens and make lower case
for word in words:
    lower_words.append(word.lower())

# Print several items from list as sanity check
lower_words[:50]
```

Remove stop words

```python
# Import nltk
import nltk
nltk.download('stopwords')

# Get English stopwords and print some of them
sw = nltk.corpus.stopwords.words('english')
sw[:10]
```

```python
# I want to get 100 words of the text that are not stop words

non_stop_words = []

for word in words:
  if word not in sw:
    non_stop_words.append(word)

print(non_stop_words[:100])
```
